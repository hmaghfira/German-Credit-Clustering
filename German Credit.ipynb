# 4) DATA MODELLING

In this section, I will use clustering algorithm such as K-Means, Hierarcical, and DBSCAN to create customer segmentation based on feature choosen.

### 4.1 K-MEANS

#### K-MEANS PROCESS

```python
# Step 1. Konversi Dataset ke Data Array 
X = np.asarray(dtnew)
```

```python
# Step 2. Lihat scatter plotnya
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(1, figsize=(8,6))
plt.clf()

ax=Axes3D(fig)
plt.cla()
ax.set_xlabel("Age")
ax.set_ylabel("Credit amount")
ax.set_zlabel("Duration")
ax.set_title("Distribution of Feature")

ax.scatter(X[:,0],X[:,1],X[:,2], label = 'True Position')

# plt.scatter(X[:,0],X[:,1], label = 'True Position')
# plt.xlabel("Credit amount")
# plt.ylabel("Duration")
# plt.title("Distribution of Feature")
```

```python
# Step 3. Tentukan K dengan elbow
Sum_of_squared_distances = []
K = range(1,15)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(X)
    Sum_of_squared_distances.append(km.inertia_)

plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
# Step 4. Melakukan aktivasi algoritma K-Means 
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
```

```python
# Step 5. Menampilkan nilai centroid yang dibangkitkan oleh algoritma kmeans 
print(kmeans.cluster_centers_)

# [[cluster, pengeluarannya]]
```

```python
print(kmeans.labels_)
```

```python
import warnings
warnings.filterwarnings('ignore')
dtnew['cluster_kmeans']=kmeans.labels_
dtnew.head()
```

```python
# Step 6 Menampilkan hasil centroid 
# Visualization of centroid
x = X[:,0]
x1 = X[:,1]
x2 = X[:,2]
group= kmeans.labels_
color = ['red','blue','black']
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
for g in set(kmeans.labels_):
    xa = [x[j] for j in range(len(x)) if group[j]==g]
    xb = [x1[j] for j in range(len(x1)) if group[j]==g]
    xc = [x2[j] for j in range(len(x1)) if group[j]==g]
    ax.scatter(xa,xb,xc, c=color[g], label=g)
    ax.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],kmeans.cluster_centers_[:,2],color='black')
ax.legend()
ax.set_xlabel("Age")
ax.set_ylabel("Credit amount")
ax.set_zlabel("Duration")
plt.title("Centroid KMeans Clustering")
plt.show()
```

```python
# Step 7. Memasukkan hasil label ke dalam suatu 
data_clustered = pd.DataFrame(X)
data_clustered['cluster'] = pd.DataFrame(kmeans.labels_)
data_clustered = data_clustered.sort_values('cluster')
data_clustered = np.asarray(data_clustered)
data_clustered
```

```python
# Dari endogram, didapatkan best cluster adalah 2
# Step 2. Melakukan fitting hierarchical clustering 
model = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
model.fit(X)
labels = model.labels_
```

### 4.3 DBSCAN

#### PROCESS OF DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
# Step 1. Plot data to scatter plot
fig = plt.figure(1, figsize=(8,6))
plt.clf()

ax=Axes3D(fig)
plt.cla()
ax.set_xlabel("Age")
ax.set_ylabel("Credit amount")
ax.set_zlabel("Duration")
ax.set_title("Distribution of Feature")

ax.scatter(X[:,0],X[:,1],X[:,2], label = 'True Position')
```

```python
Log=np.array(Logs)
Log
```

```python
# Step 3. Melakukan aktivasi algoritma DBSCAN
dbscan = DBSCAN(eps=0.25, min_samples=5)
dbscan.fit(Log)
```

```python
# Step 4. Menampilkan label dibangkitkan oleh algoritma dbscan 
print(dbscan.labels_)
```

```python
# Step 5 Menampilkan hasil centroid 
# Visualization of centroid
x = Log[:,0]
x1 = Log[:,1]
x2 = Log[:,2]
group= dbscan.labels_
color = ['red','blue','black']
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
for g in set(dbscan.labels_):
    xa = [x[j] for j in range(len(x)) if group[j]==g]
    xb = [x1[j] for j in range(len(x1)) if group[j]==g]
    xc = [x2[j] for j in range(len(x1)) if group[j]==g]
    ax.scatter(xa,xb,xc, c=color[g], label=g)
#     ax.scatter(dbscan.cluster_centers_[:,0],dbscan.cluster_centers_[:,1],dbscan.cluster_centers_[:,2],color='black')
ax.legend()
ax.set_xlabel("Age")
ax.set_ylabel("Credit amount")
ax.set_zlabel("Duration")
plt.title("DBSCAN Clustering")
plt.show()
```

# 6) DEPLOYMENT

In conclusion, I will create **macro segmentation** of customers :<br>
Type A/Cluster0 : potential customer, we should make them have higher spending.<br>
Type B/Cluster2 : satisfy customer, we should make them satisfy with our company.<br>
Type C/Cluster1 : risk customer, we should control these cluster to pay on time.<br>

For each cluster we can give this treatment :<br>
**Type A**
1. Give promotion for each purpose of credit, with longer duration.
2. Give package credit that consists most highest purpose of the cluster, with discount and longer duration.
3. Give lower interest if they take credit more than 4000 credit amount.

**Type B**
1. Give an email friendly reminder, to make customer feel happy and satisfy with our company.
2. Give promotion package of credit that consists most highest purpose of the cluster.
3. Give lower interest if they take credit lower than 6 months.


**Type C**
1. Follow up calling, as reminder to pay the credit not longer than the duration.
2. Give higher interest if they take long duration, so that they will lower their spending and duration.

Beside that, I will split each macro segmentation into micro segmentation based on their age :<br>
Young : 19 - 34 years old<br>
Adult : 35 - 49 years old<br>
Older : 50 - 75 years old<br>

Actually, this micro segmentation's purpose for creating digital marketing :<br>
**Young**: create unique digital marketing using slang word or the viral content nowadays.<br>
**Adult**: create digital marketing by showing the artist in their era.<br>
**Older**: create digital marketing by showing mini story concept with grandpas or grandmas as an actor, such as the story about grandpas and their children or grandchildren. Harapannya customer pada usia ini akan tersentuh dengan ceritanya, sehingga melakukan credit sesuai dengan yang kami iklankan.<br>
